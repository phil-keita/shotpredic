# Extremely Lossy Data Cleaning

```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('input/shot_logs.csv')
train_df.head()
```

```python
del train_df['SHOT_RESULT'] #basically FGM, but text
del train_df['W'] #basically FINAL_MARGIN, but W/L, not converting that
del train_df['CLOSEST_DEFENDER'] #it's a name, ignore for now
del train_df['player_name'] #see above
del train_df['CLOSEST_DEFENDER_PLAYER_ID'] #basically a name but in digits
del train_df['player_id'] #see above
del train_df['PTS'] #covers FGM
del train_df['MATCHUP'] #text where the only signifigant bit is covered by LOCATION
del train_df['GAME_ID'] #name but numbers again
del train_df['GAME_CLOCK'] #not a number, will probably be converted to seconds at some point, but this is a baseline so I'm ignoring

#replace all NaN in SHOT_CLOCK with 0 instead of deleting those rows
train_df['SHOT_CLOCK'] = train_df['SHOT_CLOCK'].fillna(0)
#convert LOCATION from H/A to 1/0
train_df['LOCATION'] = train_df['LOCATION'].map({'H': 1, 'A': 0})
train_df.head()
```

```python
train_df.to_csv("lossy_data.csv", index=False)
```
