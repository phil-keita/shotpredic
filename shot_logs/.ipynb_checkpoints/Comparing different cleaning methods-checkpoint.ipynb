# Prepping Sets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('cleaned_data.csv')
df_b = pd.read_csv('lossy_data.csv')
# since scikit's linear regression does not automatically handle string values
# generate a column for each defender and player on the set that HAS them
df = pd.get_dummies(df)
```

```python
# all of the non-label columns
X = df.drop('Shot_Result',axis= 1)
X.head()
```

```python
# just whether the shot was made
Y = df['Shot_Result']
Y.head()
```

```python
# all of the non-label columns
X_b = df_b.drop('FGM',axis= 1)
X_b.head()
```

```python
# just whether the shot was made
Y_b = df_b['FGM']
Y_b.head()
```

```python
# here is where we could experiment with multiple different splits
# random_state is a randomization seed to make sure the split is the same every time, can be removed/change
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42)
X_train_b, X_test_b, Y_train_b, Y_test_b = train_test_split(
    X_b, Y_b, test_size=0.3, random_state=42)
```

# Scikit Multiple Linear Regression on Cleaned Data

```python
model = LinearRegression()
model.fit(X_train,Y_train)
predictions_lr = model.predict(X_test)
predictions_lr
```

```python
print('L2 error : ', mean_squared_error(Y_test, predictions_lr))
```

**Not Bad?**

**probably should get some sort of chart to make the value visible**

```python
plt.scatter(range(len(Y_test)), Y_test)
plt.scatter(range(len(predictions_lr)), predictions_lr)
plt.show()
```

**ooh fancy plot**
wonder why so many of those are so far off the mark in one direction or the other

```python
lreg_coefficient = pd.DataFrame()
lreg_coefficient["Columns"] = X.columns
lreg_coefficient['Coefficient Estimate'] = pd.Series(model.coef_)
lreg_coefficient
```

```python
fig, ax = plt.subplots(figsize =(20, 10))

ax.bar(lreg_coefficient["Columns"],
lreg_coefficient['Coefficient Estimate'])
 
ax.spines['bottom'].set_position('zero')
 
plt.style.use('ggplot')
plt.show()
```

- for some reason this says a shot is more likely if the team is away, weird
- also the player name values are massively inflated
- applying regularization
# Scikit Ridge Regression on Cleaned (uses L2 regularization)

```python
from sklearn.linear_model import Ridge
 
# Train the model
ridgeR = Ridge(alpha = 1)
ridgeR.fit(X_train, Y_train)
predictions_rr = ridgeR.predict(X_test)
 
# calculate mean square error
print('L2 error : ', mean_squared_error(Y_test, predictions_rr))
 
# get ridge coefficient and print them
ridge_coefficient = pd.DataFrame()
ridge_coefficient["Columns"]= X_train.columns
ridge_coefficient['Coefficient Estimate'] = pd.Series(ridgeR.coef_)
ridge_coefficient
```

```python
fig, ax = plt.subplots(figsize =(20, 10))

ax.bar(ridge_coefficient["Columns"],
ridge_coefficient['Coefficient Estimate'])
 
ax.spines['bottom'].set_position('zero')
 
plt.style.use('ggplot')
plt.show()
```

```python
plt.scatter(range(len(Y_test)), Y_test)
plt.scatter(range(len(predictions_rr)), predictions_rr)
plt.show()
```

suprisingly similar despite the VERY different weights
